```python
file = '/g/lu/data/gc/lgs_metadata.fits'  # the original data?
datatable = Table.read(file, format = 'fits')
df = datatable.to_pandas()    #pandas table

bad = []
for i, row in df.iterrows():
    if row['strehl'] == -1:
        bad.append(i)
df = df.drop(bad)

#bad2 = []
#for i, row in df.iterrows():
#    if row['telemetry_file'] == None:
#        bad2.append(i)
#df_telemetry = df.drop(bad2)    # pandas table



df_tele = pd.read_csv('/u/steverobinson/work/telemetry_mjds')
df_tele.shape
len(np.unique(df_tele.file))

def find_nearest(array, value):              # This function shows whether the data in csv aligns with data in metadata.fits table
    i = (np.abs(array-value)).idxmin()       
    
    return i, array[i]


count = 0
telemetry_file = []
for i, row in df.iterrows():
    if int(row['epoch'][:2]) > 12:
        j, tele_mjd = find_nearest(df_tele.telemetry_mjd, row.mjd)
        if tele_mjd == row.mjd:
            telemetry_file.append(df_tele['file'][j])
            count += 1
        else:
            telemetry_file.append(None)
    else:
        telemetry_file.append(None)

df['telemetry_file'] = telemetry_file 
print(count)

print('Process has started.')
start_time = time.time()
count = 0
df['residual_rms'] = np.zeros(df.shape[0])
for i, row in df.iterrows():
    file = row.telemetry_file
    if file is not None:
        data = readsav(file.replace('/u/steverobinson', '/g/lu'))
        df.residual_rms[i] = data.a.RESIDUALRMS[0][0,0]
        print(df.residual_rms[i])
    count += 1
    if count % 10 == 0:
        print('Current number completed:' + str(count))
        print('Time elapsed:' + str(time.time() - start_time))
        
        
bad2 = []
for i, row in df.iterrows():
    if row['telemetry_file'] == None:
        bad2.append(i)
df = df.drop(bad2) 

t = Table.from_pandas(df)
t.write('table_with_rms.fits')   # table with rms being specified.
```

```python
def mjd_to_jd(mjd):
    """gives number of days since midnight on NOvember 17, 1858"""
    return mjd + 2400000.5

def jd_to_date(jd):
    
    
    """Convert Julian Day to date.
    
    Algorithm from 'Practical Astronomy with your Calculator or Spreadsheet', 
        4th ed., Duffet-Smith and Zwart, 2011.
    
    Parameters
    ----------
    jd : float
        Julian Day
        
    Returns
    -------
    year : int
        Year as integer. Years preceding 1 A.D. should be 0 or negative.
        The year before 1 A.D. is 0, 10 B.C. is year -9.
        
    month : int
        Month as integer, Jan = 1, Feb. = 2, etc.
    
    day : float
        Day, may contain fractional part.
        
    Examples
    --------
    Convert Julian Day 2446113.75 to year, month, and day.
    
    >>> jd_to_date(2446113.75)
    (1985, 2, 17.25)
    """
    
    jd = jd + 0.5
    
    F, I = math.modf(jd)
    I = int(I)
    
    A = math.trunc((I - 1867216.25)/36524.25)
    
    if I > 2299160:
        B = I + 1 + A - math.trunc(A / 4.)
    else:
        B = I
        
    C = B + 1524
    
    D = math.trunc((C - 122.1) / 365.25)
    
    E = math.trunc(365.25 * D)
    
    G = math.trunc((C - E) / 30.6001)
    
    day = C - E + F - math.trunc(30.6001 * G)
    
    if G < 13.5:
        month = G - 1
    else:
        month = G - 13
        
    if month > 2.5:
        year = D - 4716
    else:
        year = D - 4715
    return year, month, day

print(jd_to_date(mjd_to_jd(df.mjd[0])))
```
